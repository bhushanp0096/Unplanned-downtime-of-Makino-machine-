```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("AfterEDA_berger.csv")
data
```

```python
# Seperating input and output variables 
X=pd.DataFrame(data.iloc[:, 0:14])
y=pd.DataFrame(data.iloc[:, 14:])
X
```

```python
y
```

## TRAIN and TEST

```python
from sklearn.model_selection import train_test_split

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
X_train
```

### Model Building

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from catboost import CatBoostClassifier
```

```python
# Define a list of classifiers to test
classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    AdaBoostClassifier(),
    LogisticRegression(),
    GaussianNB(),
    SVC(),
    CatBoostClassifier()
]
```

```python
for clf in classifiers:
    clf.fit(X_train, y_train)
    
    # Make predictions on training set
    y_train_pred = clf.predict(X_train)
    # Evaluate model performance on training set
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # Make predictions on testing set
    y_test_pred = clf.predict(X_test)
    # Evaluate model performance on testing set
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    # Generate confusion matrix and classification report for testing set
    cm = confusion_matrix(y_test, y_test_pred)
    report = classification_report(y_test, y_test_pred)
    
    print(f" {clf.__class__.__name__} evaluation: ".center(100,"#"))
    print('Train accuracy: {:.2f}%'.format(train_accuracy * 100))
    print('Test accuracy: {:.2f}%'.format(test_accuracy * 100))
    print(f"Confusion Matrix:\n{cm}")
    print(f"Classification Report:\n{report}")
    print()
```

# ANN Model

```python
### random state train test split will be same with all people using random_state=16
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=16)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Storing the number of classes into the variable num_of_classes 
# as output variable being discrete in 2 categories,  
# no of neurons in output layer will be one
# activation = sigmoid
# loss = binary_crossentropy

num_of_classes = 1
```

```python
# Define the model architecture
model = Sequential()
model.add(Dense(150,input_dim = 14,activation="relu"))
model.add(Dense(200,activation="tanh"))
model.add(Dense(100,activation="tanh"))
model.add(Dense(500,activation="tanh"))
model.add(Dense(num_of_classes,activation="sigmoid"))
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Make predictions on training set
y_train_pred = (model.predict(X_train)>0.5).astype('int32')
# Evaluate model performance on training set
train_accuracy = accuracy_score(y_train, y_train_pred)

# Make predictions on testing set
y_test_pred = (model.predict(X_test)>0.5).astype('int32')
# Evaluate model performance on testing set
test_accuracy = accuracy_score(y_test, y_test_pred)

# Generate confusion matrix and classification report for testing set
cm = confusion_matrix(y_test, y_test_pred)
report = classification_report(y_test, y_test_pred)


print('Train accuracy: {:.2f}%'.format(train_accuracy * 100))
print('Test accuracy: {:.2f}%'.format(test_accuracy * 100))
print(f"Confusion Matrix:\n{cm}")
print(f"Classification Report:\n{report}")
print()
```

### Selecting the Best Model

```python
random = RandomForestClassifier(max_features=4,n_estimators=512,n_jobs=1,random_state=1,warm_start=True)
random.fit(X_train,y_train)
```

```python
train_accuracy = accuracy_score(y_train,random.predict(X_train))
test_accuracy = accuracy_score(y_test, random.predict(X_test))
train_accuracy,test_accuracy
```

```python
GB = GradientBoostingClassifier(learning_rate=0.1)
GB.fit(X_train, y_train)
```

```python
train_accuracy = accuracy_score(y_train,GB.predict(X_train))
test_accuracy = accuracy_score(y_test, GB.predict(X_test))
train_accuracy,test_accuracy
```

- from the above Random forest classifier gives the best result
